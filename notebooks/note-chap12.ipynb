## <内容まとめ>

・**ロジスティック回帰**  
分類するための回帰式を作って、分類の予測モデルを作る手法である。  
ロジスティック回帰では、教師あり学習の正解データに対して、それぞれの確率を計算して、その多寡によって最終的な分類を判断する。基本的には**確率が一番大きい結果**を優先し、分類の結果判定に用いる。  
回帰式は正解データの種類の数だけ作られる。  

・**ロジスティック回帰の実装**  
```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state = 0, C = 0.1, multi_class = 'auto', solver = 'lbfgs')
```
ここで引数Cは正規化項の重み定数の逆数である。リッジ回帰やラッソ回帰のalphaとは逆で、Cは小さいほうが、回帰式の係数を小さくしようとする働きが強くなり、過学習は防げる可能性があがる。  

・**正解データの確率を予測する**  
```python
モデル変数.predict_proba(二次元データ)
# 二次元データにはリストやデータフレームが指定できる。
```

・**ランダムフォレスト**  
ランダムフォレストは決定木の上位互換だと思っていいらしい。  
**ランダムフォレスト**は、たくさんの決定木を作成し、それぞれの木に予測させ、その結果の多数決で最終結果を求めるという手法。  
ランダムフォレストのように、さまざまな予測モデルを作成して、最終的に1つの予測結果を出す手法のことを**アンサンブル学習**という。  
ランダムフォレストでは、各決定木で利用するデータは行も列もランダムに選ぶようにして、作成する木に多様性を持たせて、多数決を行う。  
ランダムフォレストは「決定木でのバギング(後述)」を少し拡張した(列もランダム選択)手法と考えることができる。  

・**ランダムフォレストの実装**  
```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 100, random_state = 0, max_depth = 3)
# n_estimatorsで作成する決定木の数を指定
```

・**バギング**  
アンサンブル学習は**バギング**と**ブースティング**に分けることができる。 
バギングとは、訓練データを**ブートストラップサンプリング**することで、訓練データのデータセットをいくつか作り、それぞれモデルに学習させるもの。ブートストラップサンプリングとは、全データ(n個)の中から**復元抽出**により、n個のデータをランダムに抽出すること。復元抽出とは、2個以上のデータをランダムにサンプリングする際に、過去に選択したデータも選択対象に含めるサンプリング方法。  
ブートストラップサンプリングを表データに行う場合、全部でN行のデータに対して、復元抽出でN行分ランダムに選択する。1セットの表データから、様々なパターンの表データを作成することができる。作った表の数のぶんだけモデルを作成して、それぞれに学習させる。  

・**ブースティング**  
ブースティングでは、**モデルを1つずつ順番に学習させていく。** 
ブースティングでは、モデルは前のすでに学習が終わったモデルから、「ID○○番のデータを誤分類してしまった」や「学習した結果、MSE（平均2乗誤差）が△△」といった情報を受け取ることができる。モデルはこの受け取った情報を踏まえたうえで学習を進めるので、だんだんと精度が向上していく。  
ブースティングの手法にはXgboostや勾配ブースティングなどさまざまなものがある。それらの違いは、**次のモデルにどういう情報を渡すか**という点。  

・**アダブースト**  
ブースティングの手法の一つ。アダブーストでは、前回どのデータを誤分類してしまったのかという情報を渡す。各モデルは逐次的に予測性能も計算される。そして予測性能に応じて、重みを付与する。（データの受け渡しによってだんだんと予測性能があがっていく→重みが増えていく。）  

・**アダブーストの実装**  
```python
from sklearn.ensemble import AdaBoostClassifier

変数 = AdaBoostClassifier(n_estimators = モデル数, random_state = 数値, estimator = 決定木などのモデル)
```

## まとめ  

・ロジスティック回帰は分類の手法であり、正解データに分類される確率を予測する回帰式を作成する。  予測された確率が最大となるものを分類結果とする。  

・ランダムフォレストは利用するとデータと特徴量を毎回ランダムに選んでさまざまな決定木を作成する学習法。  

・ランダムフォレストの分類では、各分類木に予測をさせて、それらの多数決を行うことで、最終的な予測結論を下す。  

・ランダムフォレストは通常の決定木に比べ、過学習を防ぐことができる。  

・アダブーストは、大量のモデルを生成し、１つずつ逐次的に学習を進める。  

・アダブーストの各モデルは、前のモデルと学習結果を共有して学習を行う。学習後の予測は、各モデルの予測性能に応じて、重み付き多数決を行う。

````python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv('chap12/iris.csv')
df.head()
````

````python
# 平均値による欠損値の穴埋め
df_mean = df.mean(numeric_only=True)
train2 = df.fillna(df_mean)

x = train2.loc[:, :'花弁幅']
t = train2['種類']

# 特徴量の標準化
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
new = sc.fit_transform(x)
````

````python
x_train, x_val, y_train, y_val = train_test_split(new, t, test_size = 0.2, random_state = 0)
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state = 0, C = 0.1, multi_class = 'auto', solver = 'lbfgs')
````

````python
model.fit(x_train, y_train)
print(model.score(x_train, y_train))
model.score(x_val, y_val)
````

````python
df = pd.read_csv('chap12/Survived.csv')
df.head()
````

````python
def fill_age(df, pclass, survived, value):
    joken1 = df['Pclass'] == pclass
    joken2 = df['Survived'] == survived
    joken3 = df['Age'].isnull()
    df.loc[(joken1) & (joken2) & (joken3), 'Age'] = value
    return df
````

````python
print(df.pivot_table(index="Pclass",columns="Survived",values='Age'))

df = fill_age(df, 1, 0, 43)
df = fill_age(df, 1, 1, 35)
df = fill_age(df, 2, 0, 33)
df = fill_age(df, 2, 1, 25)
df = fill_age(df, 3, 0, 26)
df = fill_age(df, 3, 1, 20)
````

````python
col = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']

x = df[col]
t = df['Survived']


dummy = pd.get_dummies(df['Sex'], drop_first = True, dtype = int)
x = pd.concat([x, dummy], axis = 1)
x.head(2)
````

````python
from sklearn.ensemble import RandomForestClassifier
x_train, x_test, y_train, y_test = train_test_split(x, t, test_size = 0.2, random_state = 0)
model = RandomForestClassifier(n_estimators = 200, random_state = 0)
````

````python
model.fit(x_train, y_train)
print(model.score(x_train, y_train))
print(model.score(x_test, y_test))
````

````python
importance = model.feature_importances_ # 特徴量重要度
pd.Series(importance, index = x_train.columns)
````
